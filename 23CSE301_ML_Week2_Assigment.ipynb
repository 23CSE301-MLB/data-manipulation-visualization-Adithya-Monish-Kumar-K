```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset("iris")
```

```python
filtered = df[df['sepal_length'] > 5.0]
print(filtered)
```

```python
df['sepal_area'] = df['sepal_length'] * df['sepal_width']
print(df)
```

```python
grouped = df.groupby('species')['petal_length'].agg(['mean', 'median'])
print(grouped)
```

```python
sorted_df = df.sort_values(by='sepal_area', ascending=False)
print(sorted_df)
```

```python
df_missing = df.copy()
df_missing.loc[0:2, 'sepal_length'] = None
dropped = df_missing.dropna()
print(dropped)
filled = df_missing.fillna(df_missing.mean(numeric_only=True))
print(filled)
```

```python
df.to_csv('iris_with_sepal_area.csv', index=False)
print("Exported to iris_with_sepal_area.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("iris")
```

```python
sns.histplot(data=df, x="sepal_length", bins=15)
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(x="species", y="petal_length", data=df)
plt.title("Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.show()
```

```python
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=df)
plt.title("Sepal Width vs. Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend(title="Species")
plt.show()
```

```python
import numpy as np
numeric_df = df.select_dtypes(include=np.number)
corr = numeric_df.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df, hue="species", vars=["sepal_length", "sepal_width", "petal_length", "petal_width"])
plt.suptitle("Pairwise Relationships in Iris Dataset", y=1.02)
plt.show()
```
